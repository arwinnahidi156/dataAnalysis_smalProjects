```python
y=df.median_house_value.copy()
y
```

```python
X=df.drop(columns=['median_house_value','income_cat'])
X
```

```python
feat1=X.select_dtypes('float')
```

```python
feat1.info()
```

```python
import scipy.stats as stats
```

```python
feat1=feat1.apply(lambda x: stats.zscore(x))
```

```python
feat1
```

```python
pd.options.display.float_format='{:.2f}'.format
```

```python
feat1
```

```python
feat1.agg(['mean','std'])
```

```python
ocean_dummies=pd.get_dummies(X.ocean_proximity,drop_first=True)
```

```python
ocean_dummies
```

```python
X=pd.concat([feat1,ocean_dummies],axis=1)
```

```python
X
```

```python
X.info()
```

```python
X=X.astype('float')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
rfm=RandomForestRegressor(random_state=101,n_estimators=500,max_depth=75,
                         max_features='sqrt')
```

```python
rfm.fit(X_train,y_train)
```

```python
rfm.score(X_train,y_train)
```

```python
rfm2=RandomForestRegressor(random_state=101,n_estimators=500,max_depth=100,
                         max_features=10,criterion='friedman_mse')
```

```python
rfm2.fit(X_train,y_train)
```

```python
rfm2.score(X_train,y_train)
```

```python
yHat=rfm2.predict(X_test)
```

```python
yHat
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
rmse=np.sqrt(mean_squared_error(y_test,yHat))
```

```python
print(rmse)
```

```python
print(mean_absolute_error(y_test,yHat))
```
