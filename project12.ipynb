```python
sum_vs_win=pd.crosstab(olympics.Country,olympics.Edition)
sum_vs_win
```

```python
men_vs_wom=pd.crosstab(olympics.Country,olympics.Gender)
men_vs_wom
```

```python
medals_by_cats=pd.concat([sum_vs_win,men_vs_wom],axis=1)
medals_by_cats
```

```python
medals_by_cats['Total']=medals_by_cats.Summer+medals_by_cats.Winter
```

```python
medals_by_cats
```

```python
medals_by_cats.sort_values(by='Total',ascending=False,inplace=True)
```

```python
medals_by_cats
```

```python
ranks=medals_by_cats.rank(ascending=False,method='average')
ranks
```

```python
top_50=ranks.head(50)
top_50
```

```python
plt.figure(figsize=(50,10))
sns.heatmap(top_50.T,cmap='RdYlGn_r',annot=True,fmt='2.0f')
plt.show()
```
